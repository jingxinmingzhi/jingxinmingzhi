```python
import pandas as pd
```

```python
path="./ant-learn-pandas-master/datas/beijing_tianqi/beijing_tianqi_2018.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
df['bWendu']=df.loc[:,'bWendu'].str.replace("℃","").astype("int32")
```

```python
df["yWendu"]=df.loc[:,"yWendu"].str.replace("℃","").astype("int32")
```

```python
#方法一、直接赋值
#series和series做减法
df.loc[:,"Wencha"]=df["bWendu"]-df["yWendu"]
```

```python
df
```

```python
#方法二、df.apply方法
#index(axis=0),columns(axis=1)
```

```python
def get_wendu_type(x):
    if x["bWendu"]>33:
        return '高温'
    elif x["yWendu"]<-10:
        return '低温'
    else:
        return '常温'
df.loc[:,'wendu_type']=df.apply(get_wendu_type,axis=1)
    
```

```python
df['wendu_type'].value_counts()
```

```python
df.apply(get_wendu_type,axis=1)
```
