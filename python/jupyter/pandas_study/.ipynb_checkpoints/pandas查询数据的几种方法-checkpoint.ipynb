```python
#使用函数进行查询,x是某一行，依次轮询
```

```python
df.loc[lambda x:(x['bWendu']<10)&(x['yWendu']>-10)&(x["tianqi"]=='晴')&(x["aqiInfo"]=="优"),:]
```

```python
#自定义函数
def getdata(df):
    return (df['bWendu']<10)&(df['yWendu']>-10)&(df["tianqi"]=='晴')&(df["aqiInfo"]=="优")
```

```python
#函数是对象，像变量一样传递
df.loc[getdata,:]
```
